```python
# week1_day2_operators

#check that each operator comparison is true

expr = {"APP": 12.3, "MAPT": 8.5, "SNCA": 15.2}

print(expr["APP"] > 10)
print(expr["MAPT"] == 8.5)
print(expr["SNCA"] != 12.3)

#include some logical operator play. APP_high and MAPT_high are bools.
APP_high = expr["APP"] > 10        # TRUE
MAPT_high = expr["MAPT"] > 10      # FALSE

print(APP_high and MAPT_high) # False wins here because one is true and one is false, and 'and' requires both to be true 
print(APP_high or MAPT_high)  # True wins here because only one these needs to be True with an "OR" operator
print(not MAPT_high)          # 'Not' simply flips the bool to the opposite (False -> True, in this case). Output should be 'TRUE'

# Conditional statements examples (if, elif, else)
gene = "MAPT"
value = expr[gene]

if value > 12:
    print(f"{gene} is highly expressed.")
elif value > 8:
    print(f"{gene} has moderate expression.")
else:
    print(f"{gene} is lowly expressed.")

if expr["APP"] > 10 and expr["MAPT"] > 10:
    print("Both APP and MAPT are upregulated above 10 TPM.")
elif expr["APP"] > 10 or expr["MAPT"] > 10:
    print("At least one of APP or MAPT is upregulated above 10 TPM.")
else:
    print("Neither APP nor MAPT is upregulated over TPM.")

#combining these elements together in a logic check... 
expr = {"APP": 12.3, "MAPT": 8.5, "SNCA": 15.2, "GRN": 9.7, "C9orf72": 11.4}

if expr["APP"] > 10 and expr["MAPT"] > 10:
    print("Possible amyloid and tau pathway activation.")
elif expr["SNCA"] > 12:
    print("Possible alpha-synuclein aggregation pathway.")
elif expr["GRN"] < 10 and expr["C9orf72"] > 10:
    print("Possible TDP-43-related stress pathway.")
else:
    print("Expression levels within normal ranges.")
print()

#custom for loop to show each gene in the 1st expr dictionary
print (f"Gene list and levels: ")
for i, g in enumerate(expr):
    print(f"{i+1}. {g:8s} => {expr[g]:5.2f} TPM")
print()
```

```python
###############################
# Practice Challenge section for week1_day2_operators
# You’re analyzing a mock gene-expression profile from an RNA-seq experiment.
# Your goal is to automatically classify whether the sample shows patterns consistent with 
# Alzheimer’s, Parkinson’s, or frontotemporal dementia-like activity based on a few marker genes.

# Thresholds of each transcript for each disease condition
# APP > 12 and MAPT > 9        “Alzheimer’s-like expression signature”
# SNCA > 14	                   “Parkinson’s-like expression signature”
# GRN < 9 and C9orf72 > 10	   “FTD-like expression signature”
# none of the above	           “Expression pattern appears normal”

# Write a Python script that:
# 1) Prints the full dictionary (expr) nicely formatted.
# 2) Evaluates each of the conditions using if / elif / else.
# 3) Prints the classification result in a sentence
# Optionally, you can add:
# A) A print line that shows which genes were above or below the thresholds.
# B) *Multiple if statements (instead of elif) if you want to detect more than one condition at once.

#Starting Data to be checked
expr = {
    "APP": 14.2,     # amyloid precursor
    "MAPT": 9.8,     # tau
    "SNCA": 16.3,    # alpha-synuclein
    "GRN": 8.7,      # progranulin
    "C9orf72": 10.9  # repeat expansion–related 
}

print (f"Gene TPM results (2nd test data set): ")
for i, (g, v) in enumerate(expr.items(), start=1):
    print(f"{i}. {g:8s} => {v:5.1f} TPM")
print()

#Check thresholds of each gene
PosMarker = False
if expr["APP"] > 12 and expr["MAPT"] > 9:
    print("Alzheimer's-like expression signature!")
    PosMarker = True 
    print(f"APP transcripts are: {expr['APP']} (above the 12 TPM threshold).")
    print(f"MAPT transcripts are: {expr['MAPT']} (above the 9 TPM threshold).")  
    print()
if expr["SNCA"] > 14:
    print("Parkinson's-like expression signature!")
    PosMarker = True 
    print(f"SNCA transcripts are: {expr['SNCA']} (above the 14 TPM threshold).")  
    print()
if expr["GRN"] < 9 and expr["C9orf72"] > 10:
    print("FTD-like expression signature!")
    PosMarker = True 
    print(f"GRN transcripts are: {expr['GRN']} (lower than the 9 TPM threshold).")
    print(f"C9orf72 transcripts are: {expr['C9orf72']} (above the 10 TPM threshold).")  
    print()
if PosMarker == False:
    print("All expression levels are within normal ranges. Expression pattern appears normal.")
    print()




```
