```python
# basic functionality example

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Make a simple DataFrame with 2 columns
df = pd.DataFrame({
    "gene": ["APP", "MAPT", "SNCA", "GRN", "C9orf72"],
    "expression": [12.3, 8.5, 15.2, 9.7, 11.1]
})

# 2. Inspect the data
print(df.head())
print(df.info())
print(df.describe())

# 3. Simple math example to calculate and create a zscore column
df["zscore"] = (df["expression"] - df["expression"].mean()) / df["expression"].std()
print(df)

# 4. Plots 
df.plot(kind="bar", x="gene", y="expression", legend=False)
plt.title("Example Raw Expression Levels")
plt.show()

df.plot(kind="barh", x="gene", y="zscore", legend=False)
plt.title("Example Zscore Expression Levels")
plt.show()
```
