```python
# Step 5 code

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Make a simple DataFrame
df = pd.DataFrame({
    "gene": ["APP", "MAPT", "SNCA", "GRN", "C9orf72"],
    "expression": [12.3, 8.5, 15.2, 9.7, 11.1]
})

# 2. Inspect the data
print(df.head())
print(df.info())
print(df.describe())

# 3. Simple math
df["zscore"] = (df["expression"] - df["expression"].mean()) / df["expression"].std()
print(df)

# 4. Plot
df.plot(kind="barh", x="gene", y="expression", legend=False)
plt.title("Toy Expression Levels")
plt.show()
```
