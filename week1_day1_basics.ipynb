```python
# week1_day1_basics

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# 1) No variable declaration needed--python fills in types automatically
x = 10           # int
pi = 3.1415      # float
gene = "MAPT"    # string
is_neuron = True # boolean

# check variable types
print (f"x is a {type(x)} and is set to: {x}")
print (f"p1 is a {type(pi)} and is set to: {pi}")
print (f"gene is a {type(gene)} and is set to: {gene}")
print (f"is_neuron is a {type(is_neuron)} and is set to: {is_neuron}")
print()

# 2) Lists are changeable and Tuples are not changable
genes = ["APP", "MAPT", "SNCA", "GRN"]
genes.append("C9orf72")
print (f"Gene list should be: MAPT, SNCA, APP, C9orf72, GRN as called by 1,2,0,4,3 positions... and, it is:")
indicies = [1,2,0,4,3]
for i in indicies:
    print (f"{genes[i]} in index position {i}")   # MAPT, SNCA, APP, C9orf72, GRN
print ()

# 3) Dictionaries (key-value pairs, and can also be lists/arrays of pairs)
expr = {"APP":12.3, "MAPT":8.5, "SNCA":15.2}      
expr["GRN"] = 9.7           # add new key
print(f"GRN gene added as 9.7. It is: {expr['GRN']}")
print(f"All dictionary pairs: {expr.items()}")
# Test to see if dictionary entries also be called by list locations?  --no. You can loop through the list iteratively, but the index location is not numbered like a list. It is indexed by key name, not number. 
print ()

# 4) Combine these elements together to call variables with for-loops rather than individually 
gene = "MAPT"
fold_change = 1.8
print(f"{gene} shows {fold_change}-fold upregulation")
print ()

# Show list of genes using a simple for-loop, rather than calling each individually
print (f"gene list with C9orf72 appended previously:")
for g in genes:
    print(f"{g} in index {genes.index(g)}") 
print()

# Dictionary lookup using a simple for-loop, rather than calling each individually
expr = {"APP":12.3, "MAPT":8.5, "SNCA":15.2, "GRN":9.7}
print (f"gene list (keys) and values in expr dictionary:")
for k, v in expr.items():
    print(f"{k}: {v}")
#How to print the key/name of each pair? and not just the value? -- ok call the name as expr.key and the value as expr.value, and the pair as expr.items (returns both key and value)  
print()

```
