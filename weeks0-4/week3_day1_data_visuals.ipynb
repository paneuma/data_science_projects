```python
#week3_day1_data_visuals
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

# Sample data
genes = ["APP", "MAPT", "SNCA", "GRN", "C9orf72"]
fold_change = [1.4, 0.93, 0.25, 1.62, 1.05]
sd = [0.2, 0.4, 0.05, 0.02, 0.4]
df = pd.DataFrame({"Gene": genes, "Fold Change": fold_change, "sd": sd})


# Basic bar plot with error bars using pyplot
plt.bar(df["Gene"], df["Fold Change"], yerr=df["sd"], capsize=5, color="skyblue")
plt.axhline(1, color="gray", linestyle="--", linewidth=1)
plt.title("Fold Change of Gene Expression (Treated vs Control)")
plt.xlabel("Gene Name")
plt.ylabel("Fold Change (2^-Ct)")

# changing colors and adding a label 
# Savefig must come BEFORE plt.show()
plt.savefig("Gene_fold_change_bar_graph_error_bars.png", dpi=300, bbox_inches="tight")
plt.show()

#using the seaborn library to create a simple bar plot
sns.set(style="whitegrid")
sns.barplot(x="Gene", y="Fold Change", data=df, hue="Gene", legend=False, palette="coolwarm", errorbar="sd")  
# no error bars will show because there is no SD due to no replicate values in the data, but sns would automatically generate them if there were
plt.axhline(1, color="gray", linestyle="--")
plt.title("Gene Expression Fold Change (Heat Colors)")
plt.savefig("Gene_fold_change_bar_graph_heat_colors.png", dpi=300, bbox_inches="tight")
plt.show()

#Scatter plots, histograms, and regression lines in seaborn lib 
PCRdata = {
    "Gene": ["APP", "MAPT", "SNCA", "GRN", "C9orf72"],
    "Ct_Control": [23.1, 26.4, 24.9, 27.5, 29.1],
    "Ct_Treated": [21.8, 25.7, 26.2, 26.0, 27.9]
}
PCRdf = pd.DataFrame(PCRdata)

sns.scatterplot(x="Ct_Control", y="Ct_Treated", data=PCRdf, s=150, color="black")
sns.regplot(x="Ct_Control", y="Ct_Treated", data=PCRdf, color="blue")
plt.title("Treated vs Control Ct values")
plt.xlabel("Ct (Control)")
plt.ylabel("Ct (Treated)")
plt.savefig("Gene_Scatter_plot.png", dpi=300, bbox_inches="tight")


```
