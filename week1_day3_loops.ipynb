```python
# week1_day3_Loops

# starting data lists
genes = ["APP", "MAPT", "SNCA", "GRN", "C9orf72"]

expr = {
    "APP": 14.2,     # amyloid precursor
    "MAPT": 9.8,     # tau
    "SNCA": 16.3,    # alpha-synuclein
    "GRN": 8.7,      # progranulin
    "C9orf72": 10.9  # repeat expansion–related 
}
#expr is the same as previous

#g iterates through each entry in genes in order
for g in genes:
    print(f"Analyzing gene: {g}")

#i iterates from 0-4 in order (with a little formating to reduce output lines)
for i in range(5):
    if i < 4: 
        print(i, end =",")
    else:
         print(i)

#range can take up to 3 inputs (start, stop, step)
for i in range(2, 10, 2):
    print(i)
#output is 2,4,6,8 because it starts at 2, ends at 10 (stops when it hits 10, before displaying), and it increments/steps by 2

#looping through a dictionary shows all keys
for g in expr:
    print(g)
#To loop through the values of keys, use "dictionary.values()"
for v in expr.values():
    print(v)
#To loop through both keys and values, use 2 arguments in for-loop, and use "dictionary.items()"
for g, v in expr.items():
    print(f"{g}: {v}")
###### We did all this in day 2 already ... #####

#"break" ends a loop and jumps out
for g in genes:
    if g == "MAPT":
        print("Found MAPT!")
        break
# "continue" skips the current iteration and goes to the next iteration or ennumeration in the loop--it does not jump out of the loop
#in this case, when MAPT is found, it does not print, but instead goes to the next gene in the genes list 
for g in genes:
    if g == "MAPT":
        continue
    print(g)
# "pass" does nothing, but does not throw an error. Can be used while editing in-progress to avoid loop argument errors 
for g in genes:
    pass

#While loops
count = 0
while count < 5:
    print(count)
    count += 1
#always make an appropriate end conditon or the loop will run forever...
```

```python

#List comprehensions

#new_list = [[{]expression] for 'item' in 'iterable']
genes = ["APP", "MAPT", "SNCA", "GRN", "C9orf72"]
lengths = [len(g) for g in genes]
print(lengths)
#adds the string length of each key name as an int to the "length" list
#So, this is essentially like copying a list/dictionary into a new list with some transformation 

#add another condition
high_expr_genes = [g for g, v in expr.items() if v > 10]
print(high_expr_genes)
#adds key names from the expr dictionary to the "high_expr_genes" list if the expression is higher than 10
#interesting syntax uses multiple conditions (for loop and if statement), but it may be easier to understand the conditons by spreading it out a bit... 

#pairs = [(g, v) for g in genes for v in [expr[g]]] is equivalent to the constuction below. 
# THe below construction may be easier to understand quicker. Pairs is made as an empty list, and then a for loop is used to simply copy in (append) the expr keys.
pairs = []
for g in genes:
    pairs.append((g, expr[g]))
```

```python
### W1D3 Test:
# Create a script that:
# Prints all genes and expression values.
# Builds a new list containing only genes with expression > 10 TPM.
# Prints that list.
# Counts how many genes are “highly expressed.”

#creates empty list of highly expressed genes
high_expr = {}
high_count = 0

for k, v in expr.items():
    print(f"{k:8s}: {v:.1f} TMP")
    if v > 10:
        high_expr[k] = v 
        high_count +=1 
print()

print(f"The list of highly expressed genes is: ")
for l, b in high_expr.items():
    print(f"{l:8s}: {b:.1f} TMP")
print()

print(f"There are {high_count} genes in this list!")

```
